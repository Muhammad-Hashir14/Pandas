```python
import pandas as pd
```

```python
data = pd.read_csv('cast.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby([df['year'],'type']).size()
```

## UnStack

```python
c_decade = df.groupby([df['year'],'type']).size()
```

```python
#Unstack changes groupes data to dataframe
c_decade.unstack()
```

```python
graph = c_decade.unstack().head()
```

```python
graph.plot(kind = 'bar')
```

```python
decade = df.groupby(['type',df['year']]).size()
```

```python
decade
```

```python
frame = decade.unstack()
```

```python
frame
```

```python
graph = frame.iloc[:,0:9]
```

```python
graph.plot(kind = 'bar')
```
