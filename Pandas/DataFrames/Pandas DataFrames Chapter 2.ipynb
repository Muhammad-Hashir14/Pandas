## Sorting

```python
#finding all movies name Macbeth
macbeth = df[df['title'] == 'Macbeth']
macbeth
```

```python
#Now sorting movies named Macbeth by index using function sort_index
macbeth.sort_index()
```

```python
#Now sorting movies named Macbeth by year using sort_values
macbeth_year = macbeth.sort_values('year')
macbeth_year
```
